```python
#lets import our usual suspects
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# Load Data
df = pd.read_csv('flight.csv')
```

# 1 plot insights

This graph show the relationship between basic economy and total fare
Where true represents that the seat belongs to the economic class, while false represents that the seat does not belong to the economic class, so this graph show that true is less total fare because it belongs to an economy class, while false is considered higher total fare

```python
plt.figure(figsize=(15, 10))
sns.barplot(x="isBasicEconomy", y="totalFare", data=df , palette=['#F8B195','#89A4C7']).set(title='The relationship between basic economy and total Fare')
plt.xticks(rotation=90)

```

# 2 plot insights

This graph shows The relationship between total Fare and Is the flight non-stop
Where true represents that the flight is one and not stopped, while false indicates that the flight is stopped at an airport before reaching the desired destination. The drawing shows that false values are the highest total fare cost, while true is the lowest cost.

```python

plt.figure(figsize=(15, 10))
sns.barplot(x="isNonStop", y="totalFare", data=df , palette=['#a8bcd6','#F8B195']).set(title='The relationship between total Fare and Is the flight non-stop')
plt.xticks(rotation=90)

```

# 3 plot insights

This graph shows The relationship between the number of elapsed days for the flight and the total Fare with basic economy
 , The graph shows that the greater the number of elapsed days for the flight, the lower the number of people using the economy class. Of course, greater the number of elapsed days for the flight, the greater the total fare

```python

plt.figure(figsize=(15, 10))
sns.barplot(x='elapsedDays', y='totalFare', hue='isBasicEconomy', data=df, palette=['#89A4C7','#F8B195'] ).set(title='The relationship between Number of elapsed days for the flight and total Fare with basic economy')
plt.xticks(rotation=90)
```

# 4 plot insights

This graph shows The relationship between starting airport and total Fare 
The graph shows that the highest total fare is for flights departing from OAK Airport with a cost of more than 500
It is followed by SFO airport with less than 500, and in third place is LAX airport

```python

plt.figure(figsize=(15, 10))
sns.barplot(x="startingAirport", y="totalFare", data=df , palette=['#f6956f','#7494be','#F8B195','#89A4C7','#facab7','#a8bcd6']).set(title='The relationship between starting airport and total Fare')
plt.xticks(rotation=90)

```

# 5 plot insights

The graph shows that the highest total fare for the non-economy class is for flights arriving at OAK Airport, then SFO, then JFK. While the highest total fare for the economy class are the flights that arrive at OAK Airport, then SFO, then LAX

```python

plt.figure(figsize=(15, 10))
sns.barplot(x='destinationAirport', y='totalFare', hue='isBasicEconomy', data=df, palette=['#89A4C7','#F8B195']).set(title='The relationship between destination airport , basic economy and total Fare')
plt.xticks(rotation=90)
```

# 11 plot insights

This graph show Percentage of the flight non-stop

This graph represents the number of non-stop trips, where the percentage of true values represents 28% of the total values, while the false value represents 72%

```python
df_result_c = df['isNonStop'].value_counts().nlargest(4)
colors = ['#F8B195','#89A4C7']
sns.set(rc={'figure.figsize':(10,10)})
plt.pie(df_result_c.values, labels=df_result_c.index, colors=colors, autopct='%.0f%%', textprops={'fontsize': 14})
plt.title('Percentage of the flight non-stop ')
plt.show()
```
